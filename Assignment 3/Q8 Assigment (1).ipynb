```python
import selenium 
import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException,NoSuchElementException
from selenium.webdriver.common.by import By
import time
import warnings
warnings.filterwarnings('ignore')
```

```python
driver=webdriver.Chrome(r'C:\Users\admin\Downloads\chromedriver_win32\chromedriver.exe')
```

```python
driver.get('https://www.forbes.com/')
```

```python
btn = driver.find_element(By.XPATH,'//div[@class="_8FT-x3t4"]/div')
btn.click()
```

```python
Name=[]
name=driver.find_elements(By.XPATH,'//div[@class="personName"]/div')
for i in name:
    Name.append(i.text)
```

```python
Name[0:200]
```

```python
NetWorth=[]
net=driver.find_elements(By.XPATH,'//div[@class="netWorth"]/div')
for i in net:
    NetWorth.append(i.text)
```

```python
NetWorth[0:200]
```

```python
Country=[]
cout=driver.find_elements(By.XPATH,'//div[@class="countryOfCitizenship"]')
for i in cout:
    Country.append(i.text)
```

```python
Country[0:200]
```

```python
Company=[]
comp=driver.find_elements(By.XPATH,'//span[@class="source-text"]')
for i in comp:
    Company.append(i.text)
```

```python
Company[0:200]
```

```python
Rank=list(range(1,100))
```

```python
Ranks=[]
ran=driver.find_elements(By.XPATH,'//div[@class="rank"]')
for i in ran:
    Ranks.append(i.text.split('.')[0])
```

```python
Ranks[0:200]
```

```python
Industry=[]
ind=driver.find_elements(By.XPATH,'//div[@class="category"]')
for i in ind:
    Industry.append(i.text)
```

```python
Industry[0:200]
```

```python
age=[]
old=driver.find_elements(By.XPATH,'//div[@class="age"]/div')
for i in old:
    age.append(i.text)
```

```python
age[0:200]
```

```python
df=pd.DataFrame({})
df['Rank']=Ranks[0:200]
df['Name']=Name[0:200]
df['Age']=age[0:200]
df['Country']=Country[0:200]
df['Networth']=NetWorth[0:200]
df['Industry']=Industry[0:200]
df['Source']=Company[0:200]
```
