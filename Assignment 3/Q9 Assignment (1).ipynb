```python
import selenium 
import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException,NoSuchElementException
from selenium.webdriver.common.by import By
import time
import warnings
warnings.filterwarnings('ignore')
```

```python
driver=webdriver.Chrome(r'C:\Users\admin\Downloads\chromedriver_win32\chromedriver.exe')
```

```python
driver.get('https://www.youtube.com')
```

```python
driver.execute_script('window.scrollBy(0,500)')
Coment=[]
com=driver.find_elements(By.XPATH,'//yt-formatted-string[@class="style-scope ytd-comment-renderer"]')
for i in com:
    a=i.text
    Coment.append(a)
```

```python
driver.execute_script('window.scrollBy(0,500)')
Time=[]
times=driver.find_elements(By.XPATH,'//yt-formatted-string[@class="published-time-text style-scope ytd-comment-renderer"]')
for i in times:
    Time.append(i.text)
```

```python
Time[0:10]
```

```python
driver.execute_script('window.scrollBy(0,500)')
Upvote=[]
vote=driver.find_elements(By.XPATH,'//ytd-comment-action-buttons-renderer[@class="style-scope ytd-comment-renderer"]')
for i in vote:
    a=i.text.split('\n')[0]
    Upvote.append(a)
```

```python
df=pd.DataFrame({})
df['Comment']=Coment[0:500]
df['TIME']=Time[0:500]
df['Upvote']=Upvote[0:500]
```
