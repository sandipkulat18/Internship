```python
a=1
url=[]
for i in range(1,4):
    no="https://www.amazon.in/s?k=guitar&crid=5PUP1X8S748Q&sprefix=guitar%2Caps%2C736&ref=nb_sb_noss_1"+str(a)
    a=a+1
    url.append(no)
```

```python
len(url)
```

```python
brandf=[]
for i in url:
    driver.get(i)
    
    brand_1=driver.find_elements(By.XPATH,'//span[@class="a-size-base-plus a-color-base a-text-normal"]')
    for i in brand_1:
        brandf.append(i.text)
```

```python
brandf
```

```python
prod_urls = []

for i in range(3):

  

    url= driver.find_elements(By.XPATH,'//h2[@class= "a-size-mini a-spacing-none a-color-base s-line-clamp-2"]//a')

    for j in url:

        prod_urls.append(j.get_attribute('href'))

    #clicking on next button to jump on next page

    nxt_button = driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')                                         

    nxt_button.click()                                                   

    time.sleep(5)
```

```python
len(prod_urls)
```

```python
Brand_Name=[]
for i in prod_urls:

    driver.get(i)

    time.sleep(5)

    

    try:

        brand=driver.find_element(By.XPATH,"//table[@id='productDetails_techSpec_section_1']/tbody/tr[1]/td")

        Brand_Name.append(brand.text)

    except:

        Brand_Name.append('-')
```
