```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException,NoSuchElementException
from selenium.webdriver.common.by import By
import time
import warnings
warnings.filterwarnings('ignore')
```

```python
driver=webdriver.Chrome(r"‪C:\Users\admin\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
driver.get('https://www.flipkart.com/')
```

```python
search = driver.find_element(By.XPATH,"//input[@class='_3704LK']")
search.send_keys("smartphone")
```

```python
search_btn = driver.find_element(By.XPATH,"//button[@class='L0Z3Pu']")
search_btn.click()
```

```python
url_1=driver.find_elements(By.XPATH,"//a[@class='_1fQZEK']")
```

```python
url=[]
for i in url_1:
    a=i.get_attribute('href')
    url.append(a)
```

```python
brand=[]
for i in url:
    driver.get(i)
    
    brand_1=driver.find_elements(By.XPATH,"//span[@class='B_NuCI']")
    
    if len(brand_1)==0:
        brand.append("-")
        
    else:
        for j in brand_1:
            b=j.text
            c=b.split()
            d=c[0]
            brand.append(d)
```

```python
name=[]
for i in url:
    driver.get(i)
    
    name_1=driver.find_elements(By.XPATH,"//span[@class='B_NuCI']")
    
    for j in name_1:
        b=j.text
        name.append(b)
```

```python
color=[]
for i in url:
    driver.get(i)
    
    read_btn = driver.find_element(By.XPATH,"//button[@class='_2KpZ6l _1FH0tX']")
    read_btn.click()
    
    color_1=driver.find_elements(By.XPATH,"//div[@class='_1UhVsV']/div[1]/table/tbody/tr[4]/td[2]")
    
    for j in color_1:
        b=j.text
        color.append(b)
```

```python
ram=[]
for i in url:
    driver.get(i)
    
    read_btn = driver.find_element(By.XPATH,"//button[@class='_2KpZ6l _1FH0tX']")
    read_btn.click()
    
    ram_1=driver.find_elements(By.XPATH,"//div[@class='_1UhVsV']/div[4]/table/tbody/tr[2]/td[2]")
    
    for j in ram_1:
        b=j.text
        ram.append(b)
```

```python
stor=[]
for i in url:
    driver.get(i)
    
    read_btn = driver.find_element(By.XPATH,"//button[@class='_2KpZ6l _1FH0tX']")
    read_btn.click()
    
    stor_1=driver.find_elements(By.XPATH,"//div[@class='_1UhVsV']/div[4]/table/tbody/tr[1]/td[2]")
    
    for j in stor_1:
        b=j.text
        stor.append(b)
```

```python
pcam=[]
scam=[]
dis=[]
res=[]
pro=[]
proc=[]
batt=[]
price=[]

for i in url:
    driver.get(i)
    
    read_btn = driver.find_element(By.XPATH,"//button[@class='_2KpZ6l _1FH0tX']")
    read_btn.click()
    
    pcam_1=driver.find_elements(By.XPATH,"//div[@class='_1UhVsV']/div[5]/table/tbody/tr[2]/td[2]")
    scam_1=driver.find_elements(By.XPATH,"//div[@class='_1UhVsV']/div[5]/table/tbody/tr[5]/td[2]")
    dis_1=driver.find_elements(By.XPATH,"//div[@class='_1UhVsV']/div[2]/table/tbody/tr[1]/td[2]")
    res_1=driver.find_elements(By.XPATH,"//div[@class='_1UhVsV']/div[2]/table/tbody/tr[2]/td[2]")
    pro_1=driver.find_elements(By.XPATH,"//div[@class='_1UhVsV']/div[3]/table/tbody/tr[2]/td[2]")
    proc_1=driver.find_elements(By.XPATH,"//div[@class='_1UhVsV']/div[3]/table/tbody/tr[3]/td[2]")
    batt_1=driver.find_elements(By.XPATH,"//div[@class='_2418kt']/ul/li[4]")
    price_1=driver.find_elements(By.XPATH,"//div[@class='_30jeq3 _16Jk6d']")
    
    
    for j in pcam_1:
        b=j.text
        pcam.append(b)
    
    for j in scam_1:
        b=j.text
        scam.append(b)
        
    for j in dis_1:
        b=j.text
        dis.append(b)
        
    for j in res_1:
        b=j.text
        res.append(b)
        
    for j in pro_1:
        b=j.text
        pro.append(b)
        
    for j in proc_1:
        b=j.text
        proc.append(b)
        
    for j in batt_1:
        b=j.text
        batt.append(b)
        
    for j in price_1:
        b=j.text
        price.append(b)
```

```python
print(len(batt),len(price),len(brand),len(name),len(color),len(ram),len(stor),len(pcam),len(scam),len(dis),len(res),len(pro),len(proc),len(url))
```
