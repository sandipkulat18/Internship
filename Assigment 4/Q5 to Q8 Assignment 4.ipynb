```python
driver.get('https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare')
```

```python
Name=[]
Author_Name=[]
Volume_Sold=[]
Publisher=[]
Genre=[]
```

```python
name=driver.find_elements(By.XPATH,"//table[@class='in-article sortable']/tbody/tr/td[2]")
auth=driver.find_elements(By.XPATH,"//table[@class='in-article sortable']/tbody/tr/td[3]")
vol=driver.find_elements(By.XPATH,"//table[@class='in-article sortable']/tbody/tr/td[4]")
pub=driver.find_elements(By.XPATH,"//table[@class='in-article sortable']/tbody/tr/td[5]")
gen=driver.find_elements(By.XPATH,"//table[@class='in-article sortable']/tbody/tr/td[6]")
```

```python
Rank=[]
rank=driver.find_elements(By.XPATH,"//table[@class='in-article sortable']/tbody/tr/td[1]")
for i in rank:
    Rank.append(i.text)
```

```python
for i in name:
    Name.append(i.text)
for i in auth:
    Author_Name.append(i.text)
for i in vol:
    Volume_Sold.append(i.text)
for i in pub:
    Publisher.append(i.text)
for i in gen:
    Genre.append(i.text) 
```

```python
print(len(Name),len(Author_Name),len(Volume_Sold),len(Publisher),len(Genre),len(Rank))
```

```python
table=pd.DataFrame({})
table['Rank']=Rank[:]
table['Book Name']=Name[:]
table['Author Name']=Author_Name[:]
table['Volumes Sold']=Volume_Sold[:]
table['Publisher']=Publisher[:]
table['Genre']=Genre[:]
table
```

```python
driver.get('https://archive.ics.uci.edu/')
```

```python
serch=driver.find_element(By.XPATH,'/html/body/table[1]/tbody/tr/td[2]/span[2]/a')
serch.click()
```

```python
Name_d=[]
dtype=[]
d_task=[]
att_type=[]
inst_no=[]
att_no=[]
dyear=[]
```

```python
named_ds=driver.find_elements(By.XPATH,"//html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]/table/tbody/tr/td[2]/p/b/a")
dtype_ds=driver.find_elements(By.XPATH,"//html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]/p")
d_task_ds=driver.find_elements(By.XPATH,"//html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]/p")
att_type_ds=driver.find_elements(By.XPATH,"//html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]/p")
inst_no_ds=driver.find_elements(By.XPATH,"//html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]/p")
att_no_ds=driver.find_elements(By.XPATH,"//html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]/p")
dyear_ds=driver.find_elements(By.XPATH,"//html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]/p")
```

```python
for i in named_ds:
    Name_d.append(i.text)
    
for i in dtype_ds:
    dtype.append(i.text)
    
for i in d_task_ds:
    d_task.append(i.text)
    
for i in att_type_ds:
    att_type.append(i.text)
    
for i in inst_no_ds:
    inst_no.append(i.text)
    
for i in att_no_ds:
    att_no.append(i.text)
    
for i in dyear_ds:
    dyear.append(i.text)
```

```python
print(len(Name_d),len(dtype),len(d_task),len(att_type),len(inst_no),len(att_no),len(dyear))
```

```python
table=pd.DataFrame({})
table['Name']=Name_d[:]
table['Data Types']=dtype[1:]
table['Default Task']=d_task[1:]
table['Attribute types']=att_type[1:]
table['Number of Instances']=inst_no[1:]
table['Number of Attributes']=att_no[1:]
table['Year']=dyear[1:]
table
```
