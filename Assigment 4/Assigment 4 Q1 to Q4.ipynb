```python
data_1=driver.find_elements(By.XPATH,"//div[@id='main']/div[5]/div[1]/div/table/tbody/tr/td")
```

```python
rank=[]
state=[]
gsdp19=[]
gsdp18=[]
share18=[]
gdp=[]
ogsdp19=[]
ogsdp18=[]
n=1

for i in data_1:
    if n%8==1:
        rank.append(i.text)
    if n%8==2:
        state.append(i.text)
    if n%8==3:
        gsdp19.append(i.text)
    if n%8==4:
        gsdp18.append(i.text)
    if n%8==5:
        share18.append(i.text)
    if n%8==6:
        gdp.append(i.text)
    if n%8==7:
        ogsdp19.append(i.text)
    if n%8==0:
        ogsdp18.append(i.text)
    n=n+1
```

```python
print(len(rank),len(state),len(gsdp19),len(gsdp18),len(share18),len(gdp),len(ogsdp19),len(ogsdp18))
```

```python
table=pd.DataFrame({})
table['Rank']=rank[:]
table['State']=state[:]
table['GSDP 19-20(Cr INR at current prices)']=gsdp19[:]
table['GSDP 18-19(Cr INR at current prices)']=gsdp18[:]
table['Share(18-19)']=share18[:]
table['GDP($ billion)']=gdp[:]
table['GSDP 19-20(Cr INR at 2011-12 prices)']=ogsdp19[:]
table['GSDP 18-19(Cr INR at 2011-12 prices']=ogsdp18[:]
table
```

```python
Language=[]
contri=[]
Descp=[]
Lang=[]
star=[]
Title=[]
```

```python
repository =driver.find_elements(By.XPATH,"//h1[@class='h3 lh-condensed']/a")
repo_urls=[]
for i in repository[0:30]:
    a=i.get_attribute("href")
    repo_urls.append(a)

for i in repo_urls:

    driver.get(i)

    time.sleep(3)
    #Scraping Languages-
    try:

        cont_tags = driver.find_element(By.XPATH,'//span[@class="color-fg-default text-bold mr-1"]')

        Language.append(cont_tags.text)

    except NoSuchElementException:

        Language.append('-')
    #Scraping Languages-
    try:
        contri_1=driver.find_elements(By.XPATH,"//div[@class='BorderGrid BorderGrid--spacious']/div[5]/div/h2/a/span")
        if len(contri_1)==0:
            contri_2=driver.find_elements(By.XPATH,"//div[@class='BorderGrid BorderGrid--spacious']/div[4]/div/h2/a/span")
        
            if len(contri_2)==1:
                for k in contri_2:
                    contri.append(k.text)
            else:
                contri.append('-')
            
        else:
            for j in contri_1:
                contri.append(j.text)
    except NoSuchElementException:
        contri.append('-')
      
   
      #Scraping Languages-      
    try:
        star_1=driver.find_elements(By.XPATH,'//span[@class="Counter js-social-count"]')
        for i in star_1:
            star.append(i.text)

    except NoSuchElementException:
        star.append('-')
    #Scraping Languages-
    try:
        title=driver.find_elements(By.XPATH,'//div[@class="flex-auto min-width-0 width-fit mr-3"]')
        for i in title:
            Title.append(i.text)
        
    except NoSuchElementException:
        
        Title.append('-')
    try:
        descrip_1=driver.find_element(By.XPATH,'//p[@class="f4 my-3"]')
        Descp.append(descrip_1.text)
    except NoSuchElementException:
        Descp.append('-')
 

        
```

```python
print(len(Title),len(Descp),len(Language),len(contri),len(star))
```

```python
table=pd.DataFrame({})
table['Repository Title']=Title[:]
table['Repository Description']=Descp[:]
table['Language Used']=Language[:]
table['Contributors count)']=contri[:]
table['Stars']=star[:]
table
```
