```python
from bs4 import BeautifulSoup
import requests
```

```python
url=requests.get('https://www.cnbc.com/world/?region=world')
```

```python
url
```

```python
soup=BeautifulSoup(url.text,'html.parser')
```

```python

Time=[]
News_link=[]
```

```python
news=soup.find_all('div',class_="LatestNews-container")
for i in news:
    name=i.a.text
    News_link.append(name)
```

```python
News_link







```

```python
time=soup.find_all('time',class_="LatestNews-timestamp")
```

```python
for i in time:
    tim=i.get_text()
    Time.append(tim)
    
```

```python
Time
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({})
df['Breaking New']=News_link
df['Time']=Time
```

```python
df
```

```python
Headline=[]

for i in soup.find_all('div',class_='LatestNews-headlineWrapper'):

    l=i.a.get("href")

    Headline.append(l)
```
