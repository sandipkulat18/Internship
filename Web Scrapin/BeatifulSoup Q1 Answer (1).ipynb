```python
soup=BeautifulSoup(url.text,'html.parser')
```

```python
soup
```

```python
Headers=soup.find_all(['h1','h2','h3','h4','h5'])

print('List of the All Headers:',Headers,sep='\n\n')
```

```python
Source=[]
for i in Headers:
    Source.append(i.text)
    
print(Source)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({"Heading":Source})
```
