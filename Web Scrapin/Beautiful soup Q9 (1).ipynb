# Write a python program to scrape mentioned details from dineout.co.in :

```python
from bs4 import BeautifulSoup
import requests
```

```python
url=requests.get('https://www.dineout.co.in/delhi-restaurants/welcome-back')
```

```python
url.content
```

```python
soup=BeautifulSoup(url.content)
```

```python
Restaurant_name=[]
Cuisine=[]
Location=[]
Ratings=[]
Image_URL=[]
title=[]
```

```python
Name=soup.find_all('a',class_="restnt-name ellipsis")
for i in Name:
    Restaurant_name.append(i.text)
```

```python
Restaurant_name
```

```python
location=soup.find_all('div',class_="restnt-loc ellipsis")
    
```

```python
for i in location:
    Location.append(i.text)
```

```python
Location
```

```python
rating=soup.find_all('div',class_="restnt-rating rating-4")
for i in rating:
    Ratings.append(i.text)
```

```python
Ratings
```

```python
images=soup.find_all('img',class_="no-img")
```

```python
for i in images:
    Image_URL.append(i.get('data-src'))
```

```python
Image_URL
```

```python
cuis=soup.find_all("span", class_="double-line-ellipsis")
```

```python
for i in cuis:
    Cuisine.append(i.text.split('|')[1])
```

```python
Cuisine
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({})
df['Restaurant_name']=Restaurant_name
df['Ratings']=Ratings
df['Location']=Location
df['Image_URL']=Image_URL
df['Cuisine']=Cuisine
```
