```python
from bs4 import BeautifulSoup
import requests
```

```python
url=requests.get('https://en.wikipedia.org/wiki/Main_Page')
```

```python
url
```

```python
soup=BeautifulSoup(url.content)
soup
```

```python
Header=soup.find_all(['h1', 'h2','h3','h4','h5','h6'])
print('List all the header tags :', Header, sep='\n\n')
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({'Header':Header})
```
