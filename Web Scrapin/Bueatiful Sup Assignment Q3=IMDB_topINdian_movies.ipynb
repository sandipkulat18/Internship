3.Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of
release) and make data frame.

```python
import requests 
from bs4 import BeautifulSoup
```

```python
url=requests.get('https://www.imdb.com/india/top-rated-indian-movies')
url
```

```python
url.content
```

```python
soup=BeautifulSoup(url.text, 'html.parser')
```

```python
Movie_Name=[]
IMDB_Rating=[]
Year_of_Release=[]
```

```python
Name=soup.find_all('td',class_="titleColumn")
for i in Name:
    title =i.a.text
    Movie_Name.append(title)
    year = i.span.text.replace('(','')
    year_2=year.replace(')','')
    Year_of_Release.append(year_2)
```

```python
Movie_Name[0:12]
```

```python
rating = soup.find_all('td', class_="ratingColumn imdbRating")
for i in rating:
    imdb=i.strong.text
    IMDB_Rating.append(imdb)
```

```python
IMDB_Rating[0:10]
    
```

```python
Year_of_Release[0:12]
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({})
df['Movies_Name']=Movie_Name[0:100]
df['Year Of Release']=Year_of_Release[0:100]
df['IMDB RAting']=IMDB_Rating[0:100]
```
