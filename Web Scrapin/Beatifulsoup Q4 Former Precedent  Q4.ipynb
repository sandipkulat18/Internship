```python
import  requests
from  bs4 import BeautifulSoup
```

```python
link=requests.get('https://presidentofindia.nic.in/former-presidents.htm')
```

```python
link
```

```python
soup=BeautifulSoup(link.content)
```

```python
soup
```

```python
soup=BeautifulSoup(link.text)
```

```python
soup
```

```python
Name=[]
info=soup.find_all(['h3'])
for i in info:
    Name.append(i.text)
print(Name)
```

```python
import pandas as pd
df=pd.DataFrame({'Name Of Precident':Name})
```

```python
df
```

```python
Term=[]
for i in soup.find_all('div',class_="presidentListing"):  

    Term.append(i.p.text.split(',')[1:2])
         
    

Term
```
